# 1: What is hypothesis testing in statistics?

### Hypothesis Testing in Statistics

Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data.

**Steps of Hypothesis Testing:**
1. **Set up hypotheses**
   - Null hypothesis (H₀): assumes no effect or no difference.
   - Alternative hypothesis (H₁): assumes there is an effect or difference.

2. **Choose significance level (α)**
   - Common choice is 0.05 (5%).

3. **Collect data and calculate test statistic**
   - Example: z-score, t-score, chi-square value.

4. **Find p-value or critical value**
   - p-value shows the probability of observing the sample result if H₀ is true.

5. **Decision rule**
   - If p-value ≤ α → Reject H₀ (evidence supports H₁).
   - If p-value > α → Fail to reject H₀ (not enough evidence against H₀).



# 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

### Null Hypothesis vs Alternative Hypothesis

**Null Hypothesis (H₀):**  
- It is the default assumption.  
- States that there is **no effect**, **no relationship**, or **no difference**.  
- Example: "The mean weight of apples is 100g."

**Alternative Hypothesis (H₁ or Ha):**  
- It is what we want to test or prove.  
- States that there **is an effect**, **a relationship**, or **a difference**.  
- Example: "The mean weight of apples is not 100g."

**Key Difference:**  
- H₀ assumes *status quo* (nothing changes).  
- H₁ assumes there is a *real change or effect*.  
- In hypothesis testing, we use sample data to decide whether to reject H₀ in favor of H₁.


# 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

### Significance Level in Hypothesis Testing

**Significance Level (α):**  
- It is the probability of making a **Type I error** (rejecting the null hypothesis when it is actually true).  
- Common choices: 0.05 (5%), 0.01 (1%), or 0.10 (10%).  

**Role in Hypothesis Testing:**  
1. α is set **before** the test begins.  
2. It defines the cutoff for deciding whether results are "statistically significant."  
3. If **p-value ≤ α**, we **reject H₀** (evidence supports H₁).  
4. If **p-value > α**, we **fail to reject H₀** (not enough evidence).  

**Example:**  
- If α = 0.05, it means we accept a 5% chance of wrongly rejecting H₀.  
- So, α acts like a "risk tolerance" in hypothesis testing.


# 4: What are Type I and Type II errors? Give examples of each.

### Type I and Type II Errors

**Type I Error (False Positive):**  
- Rejecting the null hypothesis (H₀) when it is actually true.  
- Probability of this error = significance level (α).  
- Example: A medical test says a healthy person has a disease.

**Type II Error (False Negative):**  
- Failing to reject the null hypothesis (H₀) when it is actually false.  
- Probability of this error = β (beta).  
- Example: A medical test fails to detect a disease in a sick person.

**Summary Table:**

| Decision \ Reality       | H₀ True                     | H₀ False                     |
|---------------------------|-----------------------------|-------------------------------|
| Reject H₀                | Type I Error (False Positive) | Correct Decision (Reject False H₀) |
| Fail to Reject H₀        | Correct Decision (Keep True H₀) | Type II Error (False Negative) |



# 5: What is the difference between a Z-test and a T-test? Explain when to use each.

### Difference Between Z-test and T-test

**Z-test:**  
- Used when the **population variance (σ²) is known**.  
- Sample size is **large (n > 30)**.  
- Based on the **standard normal distribution (Z-distribution)**.  
- Example: Testing if the average height of 1000 students is 160 cm when population variance is known.

**T-test:**  
- Used when the **population variance is unknown**.  
- Sample size is **small (n ≤ 30)**.  
- Based on the **t-distribution**, which is wider to account for extra uncertainty.  
- Example: Testing if the average marks of 20 students differ from 50 when variance is unknown.

**Key Difference:**  
- Z-test → Large sample, variance known.  
- T-test → Small sample, variance unknown.

```python
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

population_mean = 50   
alpha = 0.05           


sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)


z_stat = (sample_mean - population_mean) / (sample_std / np.sqrt(n))


p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)


if p_value < alpha:
    print("Reject H0: There is significant difference from population mean.")
else:
    print("Fail to reject H0: No significant difference from population mean.")
```

# 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42) 
data = np.random.normal(loc=50, scale=5, size=100)  


sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
n = len(data)


alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha/2)  # z* for 95%
margin_of_error = z_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print("Sample Mean:", sample_mean)
print("95% Confidence Interval: ({:.3f}, {:.3f})".format(ci_lower, ci_upper))


plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"Lower CI = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"Upper CI = {ci_upper:.2f}")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

# 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt


def calculate_zscores(data):
    mean = np.mean(data)
    std = np.std(data, ddof=1)  
    z_scores = [(x - mean) / std for x in data]
    return z_scores


data = [50, 52, 48, 47, 53, 55, 49, 51, 50, 54]

z_scores = calculate_zscores(data)

print("Original Data:", data)
print("Z-scores:", z_scores)

plt.hist(z_scores, bins=5, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.show()
```
